```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:/datasets/Electric_Vehicle_Population_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
ev_sales=df['Model Year'].value_counts()
sns.barplot(x=ev_sales.index,y=ev_sales.values,palette='viridis')
plt.xticks(rotation='vertical')
plt.xlabel('Model Year')
plt.ylabel('Total  EV sales')
plt.show()
```

```python
ev_country=df['County'].value_counts()
ev_country
```

```python
ev_country=ev_country.head(10)
plt.bar(x=ev_country.index,height=ev_country.values,color=['r','g','b','y'])
plt.xticks(rotation='vertical')
plt.xlabel('Country registered')
plt.ylabel('Total EV counts')
plt.show()
```

```python
ev_type=df['Electric Vehicle Type'].value_counts()
sns.barplot(x=ev_type.index,y=ev_type.values,palette='viridis')
plt.xticks(rotation='vertical')
plt.title('EV type distribution')
plt.show()
```

```python
car_maker=df['Make'].value_counts().head(15)
sns.barplot(y=car_maker.index,x=car_maker.values,palette='cubehelix')
plt.show()
```

```python
top_3_makes = car_maker.head(3).index

# filtering the dataset for these top manufacturers
top_makes_data = df[df['Make'].isin(top_3_makes)]
ev_model_distribution_top_makes = top_makes_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
top_models = ev_model_distribution_top_makes.head(10)

plt.figure(figsize=(8,6))
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="viridis")
plt.title('Top Models in Top 3 Makes by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Electric Range'], bins=30, kde=True, color='royalblue')
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Number of Vehicles')
plt.axvline(df['Electric Range'].mean(), color='red', linestyle='--', label=f'Mean Range: {df["Electric Range"].mean():.2f} miles')
plt.legend()
plt.show()
```

```python
average_range_by_model = top_makes_data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()

# the top 10 models with the highest average electric range
top_range_models = average_range_by_model.head(10)

plt.figure(figsize=(8,6))
barplot = sns.barplot(x='Electric Range', y='Model', hue='Make', data=top_range_models, palette="cool")
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.show()
```
